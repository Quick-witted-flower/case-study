```python
# Definicja zmiennych
PV = 120000  # Obecna cena mieszkania w złotych
r = 0.05     # Roczna stopa wzrostu (5%)
n = 5        # Liczba lat

# Obliczenie przyszłej wartości
FV = PV * (1 + r) ** n
print("Przewidywana cena mieszkania za 5 lat:", FV)
```

Przewidywana cena mieszkania za 5 lat to 153,254 zł.

```python
import numpy_financial as npf

# Obliczenie miesięcznej wpłaty przy użyciu numpy-financial
FV = 153254       # Przyszła wartość (cena mieszkania)
i = 0.01          # Miesięczna stopa procentowa
n = 60            # Liczba miesięcy (5 lat)

miesieczna_wplata = npf.pmt(rate=i, nper=n, pv=0, fv=-FV)
print("Miesięczna wpłata potrzebna do uzbierania na mieszkanie:", miesieczna_wplata)
```

Miesięczna wpłata potrzebna do uzbierania na mieszkanie w ciągu 5 lat wynosi około 1,877 zł (zaokrąglone).

```python
import numpy as np
import numpy_financial as npf
import matplotlib.pyplot as plt

PV_mieszkanie = 120000  # Obecna cena mieszkania
stopa_rocznego_wzrostu = 0.05  
FV_mieszkanie = 153254  # Przyszła cena mieszkania po 5 latach
miesieczna_stopa_procentowa = 0.01  
liczba_miesiecy = 60  # Liczba miesięcy (5 lat)

miesieczna_stop_wzrostu_mieszkania = (1 + stopa_rocznego_wzrostu) ** (1 / 12)
miesiace = np.arange(liczba_miesiecy)
ceny_mieszkania = PV_mieszkanie * (miesieczna_stop_wzrostu_mieszkania ** miesiace)

miesieczna_wplata = npf.pmt(rate=miesieczna_stopa_procentowa, nper=liczba_miesiecy, pv=0, fv=-FV_mieszkanie)

wartosc_lokaty = npf.fv(rate=miesieczna_stopa_procentowa, nper=miesiace, pmt=-miesieczna_wplata, pv=0)

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
plt.plot(miesiace, ceny_mieszkania, label="Cena mieszkania", linestyle='-', linewidth=2)
plt.plot(miesiace, wartosc_lokaty, label="Wartość lokaty", linestyle='--', linewidth=2)
plt.xlabel("Miesiące")
plt.ylabel("Kwota (zł)")
plt.title("Prognozowana cena mieszkania i wartość lokaty")
plt.legend()
plt.grid(True)
plt.show()
```
