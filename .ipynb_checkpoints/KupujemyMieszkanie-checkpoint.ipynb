```python
# Definicja zmiennych
PV = 120000  # Obecna cena mieszkania w złotych
r = 0.05     # Roczna stopa wzrostu (5%)
n = 5        # Liczba lat

# Obliczenie przyszłej wartości
FV = PV * (1 + r) ** n
print("Przewidywana cena mieszkania za 5 lat:", FV)
```

Przewidywana cena mieszkania za 5 lat to 153,254 zł.

```python
import numpy_financial as npf

# Obliczenie miesięcznej wpłaty przy użyciu numpy-financial
FV = 153254       # Przyszła wartość (cena mieszkania)
i = 0.01          # Miesięczna stopa procentowa
n = 60            # Liczba miesięcy (5 lat)

miesieczna_wplata = npf.pmt(rate=i, nper=n, pv=0, fv=-FV)
print("Miesięczna wpłata potrzebna do uzbierania na mieszkanie:", miesieczna_wplata)
```

Miesięczna wpłata potrzebna do uzbierania na mieszkanie w ciągu 5 lat wynosi około 1,877 zł (zaokrąglone).

```python
import numpy as np
import matplotlib.pyplot as plt

# Definicja zmiennych
PV_mieszkanie = 120000  # Obecna cena mieszkania
stopa_rocznego_wzrostu = 0.05  # Roczna stopa wzrostu ceny mieszkania
FV_mieszkanie = 153254  # Przyszła cena mieszkania po 5 latach
miesieczna_stopa_procentowa = 0.01  # Miesięczna stopa procentowa lokaty
liczba_miesiecy = 60  # Liczba miesięcy (5 lat)
miesieczna_wplata = 1876.51  # Miesięczna wpłata

# Obliczenie miesięcznego wzrostu ceny mieszkania
miesieczna_stop_wzrostu_mieszkania = (1 + stopa_rocznego_wzrostu) ** (1 / 12)
ceny_mieszkania = [PV_mieszkanie * (miesieczna_stop_wzrostu_mieszkania ** i) for i in range(liczba_miesiecy)]

# Obliczenie wartości lokaty na koniec każdego miesiąca
wartosc_lokaty = []
aktualna_wartosc_lokaty = 0

for i in range(liczba_miesiecy):
  
    aktualna_wartosc_lokaty = aktualna_wartosc_lokaty * (1 + miesieczna_stopa_procentowa) + miesieczna_wplata
    wartosc_lokaty.append(aktualna_wartosc_lokaty)

# Tworzenie wykresu
plt.figure(figsize=(10, 6))
plt.plot(range(liczba_miesiecy), ceny_mieszkania, label="Cena mieszkania", linestyle='-', linewidth=2)
plt.plot(range(liczba_miesiecy), wartosc_lokaty, label="Wartość lokaty", linestyle='--', linewidth=2)
plt.xlabel("Miesiące")
plt.ylabel("Kwota (zł)")
plt.title("Prognozowana cena mieszkania i wartość lokaty")
plt.legend()
plt.grid(True)
plt.show()
```
